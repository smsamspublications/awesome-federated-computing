EDA for numerical data involves understanding the distribution, central tendency, spread, and identifying potential outliers. Here are some Python commands with their results, demonstrating these concepts using a sample dataset:

First, let's set up a synthetic dataset using pandas and numpy to illustrate these concepts.

```python
import warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set a random seed for reproducibility
np.random.seed(42)

# 1. Generate Sample Data
data = {
    'Age': np.random.normal(loc=35, scale=10, size=100).astype(int),
    'Income': np.random.normal(loc=50000, scale=15000, size=100),
    'Experience_Years': np.random.randint(0, 40, size=100),
    'Score': np.random.uniform(50, 100, size=100)
}
df = pd.DataFrame(data)

# Introduce some missing values and outliers for demonstration
df.loc[df.sample(frac=0.05).index, 'Income'] = np.nan
df.loc[np.random.randint(0, 100, size=3), 'Age'] = [75, 80, 10] # Outliers
df.loc[np.random.randint(0, 100, size=2), 'Income'] = [150000, 5000] # Outliers

print("--- Initial DataFrame (First 5 Rows) ---")
print(df.head())
print("\n")
```

2. Descriptive Statistics
The describe() method provides a quick summary of the central tendency, dispersion, and shape of a dataset's distribution, excluding NaN values.

```python
print("--- Descriptive Statistics ---")
print(df.describe())
print("\n")
```

Data Types and Missing Values
info() provides a concise summary of a DataFrame, including data types and non-null values. isnull().sum() helps identify missing values in each column.

```python
print("--- DataFrame Info (Data Types and Non-Null Counts) ---")
df.info()
print("\n")

print("--- Missing Values Count ---")
print(df.isnull().sum())
print("\n")
```

4. Distribution Visualization (Histograms)
Histograms display the distribution of numerical data. The kde=True argument adds a Kernel Density Estimate line, which is a smoothed representation of the distribution.

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(df['Age'].dropna(), kde=True, bins=15)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
sns.histplot(df['Income'].dropna(), kde=True, bins=15)
plt.title('Distribution of Income')
plt.xlabel('Income')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
sns.histplot(df['Experience_Years'].dropna(), kde=True, bins=15)
plt.title('Distribution of Experience Years')
plt.xlabel('Experience Years')
plt.ylabel('Frequency')

plt.tight_layout()
plt.savefig('numerical_data_histograms.png')
print("Generated 'numerical_data_histograms.png' for distributions.\n")
```

5. Distribution Visualization (Box Plots)
Box plots are excellent for visualizing the distribution of numerical data and identifying potential outliers. They show the median, quartiles (25th and 75th percentiles), and potential outliers.

```python
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.boxplot(y=df['Age'].dropna())
plt.title('Box Plot of Age')
plt.ylabel('Age')

plt.subplot(1, 3, 2)
sns.boxplot(y=df['Income'].dropna())
plt.title('Box Plot of Income')
plt.ylabel('Income')

plt.subplot(1, 3, 3)
sns.boxplot(y=df['Experience_Years'].dropna())
plt.title('Box Plot of Experience Years')
plt.ylabel('Experience Years')

plt.tight_layout()
plt.savefig('numerical_data_boxplots.png')
print("Generated 'numerical_data_boxplots.png' for outliers.\n")
```

6. Outlier Detection (using IQR method)
The Interquartile Range (IQR) method is a common way to detect outliers. Values that fall below Q1 - 1.5 * IQR or above Q3 + 1.5 * IQR are considered outliers.

```python
# 6. Outlier Detection (using IQR for 'Income' column)
Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_income = df[(df['Income'] < lower_bound) | (df['Income'] > upper_bound)]

print("--- Outlier Detection (Income Column using IQR) ---")
print(f"Q1 (25th percentile): {Q1:.2f}")
print(f"Q3 (75th percentile): {Q3:.2f}")
print(f"IQR (Interquartile Range): {IQR:.2f}")
print(f"Lower Bound for Outliers: {lower_bound:.2f}")
print(f"Upper Bound for Outliers: {upper_bound:.2f}")
print("\nPotential Outliers in Income:")
if not outliers_income.empty:
    print(outliers_income[['Income']])
else:
    print("No significant outliers detected in Income based on IQR method.")
```

**5. Broadcasting**
Concept: Broadcasting is a powerful NumPy feature that allows operations between arrays of different shapes, provided they are compatible. This is heavily used in data science for efficient computations without explicit looping.

```python
arr_2d = np.array([[1, 2, 3], [4, 5, 6]])
arr_1d = np.array([10, 20, 30])

print("2D Array:\n", arr_2d)
print("1D Array:", arr_1d)

# Adding a 1D array to a 2D array (broadcasting)
print("\n2D Array + 1D Array (Broadcasting):\n", arr_2d + arr_1d)
```

```python
# Example with a column vector
col_vector = np.array([[10], [20]])
print("\nColumn Vector:\n", col_vector)

matrix = np.array([[1, 2, 3], [4, 5, 6]])
print("Matrix:\n", matrix)

print("Matrix + Column Vector (Broadcasting):\n", matrix + col_vector)
```

**6. Linear Algebra Operations**
Concept: Many data science algorithms (e.g., PCA, regression) heavily rely on linear algebra. NumPy's linalg module provides robust functions.

```python
# Inverse of a matrix
A = np.array([[1, 2], [3, 4]])
print("Matrix A:\n", A)
print("Inverse of A:\n", np.linalg.inv(A))
```

```python
# Determinant of a matrix: The determinant of a matrix AA is denoted as det(A)det(A) or ∣A∣∣A∣. It is defined only for square matrices (matrices with the same number of rows and columns). The determinant serves as a scaling factor for transformations represented by the matrix and is crucial in various applications, including solving systems of linear equations and finding the inverse of a matrix
print("Determinant of A:", np.linalg.det(A)) # Expected: -2.0
```

```python
# Eigenvalues and eigenvectors (crucial for PCA)
eigenvalues, eigenvectors = np.linalg.eig(A)
print("\nEigenvalues of A:", eigenvalues)
print("Eigenvectors of A:\n", eigenvectors)
```
